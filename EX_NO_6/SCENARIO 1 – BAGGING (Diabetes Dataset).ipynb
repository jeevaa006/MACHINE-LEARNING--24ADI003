```python
# Roll No: 24BAD047

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
df = pd.read_csv(r"C:\Users\Admin\Desktop\4_sem\ML\EX_NO_6\diabetes_bagging.csv")
```

```python
X = df.drop("Outcome", axis=1)
y = df["Outcome"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)
```

```python
bag = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=50)
bag.fit(X_train, y_train)
bag_pred = bag.predict(X_test)
```

```python
dt_acc = accuracy_score(y_test, dt_pred)
bag_acc = accuracy_score(y_test, bag_pred)
```

```python
print("Decision Tree Accuracy:", dt_acc)
print("Bagging Accuracy:", bag_acc)
```

```python
plt.figure()
plt.bar(["Decision Tree", "Bagging"], [dt_acc, bag_acc])
plt.title("Accuracy Comparison")
plt.show()
```

```python
cm = confusion_matrix(y_test, bag_pred)
plt.figure()
plt.imshow(cm)
plt.title("Confusion Matrix")
plt.colorbar()
plt.show()
```
